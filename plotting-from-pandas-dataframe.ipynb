```python
#import
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#read dataframe
hearth_disease = pd.read_csv("heart-disease.csv")
hearth_disease.head()
```

```python
# find people over 50 years old age
over_50 = hearth_disease[hearth_disease["age"]>50]
over_50.head()
```

```python
#create subplot
fig , (ax0,ax1) = plt.subplots(figsize = (15,10),
                       nrows = 2,
                       ncols = 1)
```

```python
plt.style.use('seaborn-darkgrid')

fig , (ax0,ax1) = plt.subplots(figsize = (15,10),
                       nrows = 2,
                       ncols = 1,
                        sharex = True)

# data into charts
scatter = ax0.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c= over_50["target"],
                     cmap = "winter");
ax0.set(ylabel = "cholesterol",
       title = "Cholesterol-age charts")
ax0.legend(*scatter.legend_elements(),title = "target")
ax0.axhline(over_50["chol"].mean(),
            linestyle = "--")

scatter = ax1.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"],
                     cmap = "plasma")

ax1.set_xlim([50,80])
ax1.set_ylim([60,198])

ax1.legend(*scatter.legend_elements(),title = "target")
ax1.axhline(over_50["thalach"].mean(),linestyle = ":")
ax1.set(xlabel = "age",
       ylabel = "thalach",
       title = "hearth rate-age charts")

fig.suptitle("Hearth disease analysis",fontweight = "bold");
```
